```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Annual Emissions Data

```python
# load 2022 annual emissions data for coal, gas, oil, and other plants 
coal = pd.read_csv("../data/annual_emissions/annual-emissions-2022-coal.csv")
coal['type'] = "1 coal"

gas = pd.read_csv("../data/annual_emissions/annual-emissions-2022-gas.csv")
gas['type'] = "3 gas"

oil = pd.read_csv("../data/annual_emissions/annual-emissions-2022-oil.csv")
oil['type'] = "2 oil"

other = pd.read_csv("../data/annual_emissions/annual-emissions-2022-other.csv")
other['type'] = "4 other"

data = pd.concat([coal, gas, oil, other], ignore_index=True)
```

### Data Processing

```python
# coal type for conversion 
coalType = pd.read_csv("../data/other/coal-type.csv")

# keep relevant variables and change units 
data = data.merge(coalType, on=['State', 'Facility Name', 'Facility ID'], how='left')

data = data.rename(columns={'State': 'state',
 'Facility Name': 'name',
 'Facility ID': 'id',
 'Heat Input (mmBtu)': 'heatInput',
 'Gross Load (MWh)': 'grossLoad',
 'Steam Load (1000 lb)': 'steamLoad',
 'SO2 Mass (short tons)': 'so2Mass',
 'CO2 Mass (short tons)': 'co2Mass',
 'NOx Mass (short tons)': 'noxMass',
 'Fuel Type': 'coalType',
 'Energy': 'energy'})

# keep relevant variables and change units 
data['steamInd'] = 0
data.loc[(~data['steamLoad'].isna() & data['grossLoad'].isna()), 'steamInd'] = 1

data['energyConvert'] = np.nan
data.loc[(data['steamInd'] == 1) & (data['energy'].notnull()), 'energyConvert'] = data['energy']
data.loc[(data['steamInd'] == 1) & (data['energy'].notnull()) & (data['type'] == "1 coal"), 'energyConvert'] = 9845
data.loc[(data['steamInd'] == 1) & (data['energy'].notnull()) & (data['type'].isin(["2 oil", "3 gas"])), 'energyConvert'] = 19750

data['load'] = data['grossLoad']
data.loc[data['grossLoad'].isna(), 'load'] = data['steamLoad']

data['co2Mass'] = data['co2Mass'] * 0.907185 * 1000
data['so2Mass'] = data['so2Mass'] * 0.907185 * 1000
data['noxMass'] = data['noxMass'] * 0.907185 * 1000

data = data[['state', 'name', 'id', 'type', 'heatInput', 'load', 'co2Mass', 'so2Mass', 'noxMass']]

# keep observations with no missing heat input and load 
data = data.dropna(subset=['heatInput', 'load'])
data = data.sort_values(by=['id', 'type'])
data = data[data['load'] != 0]
```

```python
# check whether separate data for separate types 
dataFacility = data.groupby(['state', 'name', 'id'])['heatInput'].mean().reset_index()
data = pd.merge(data, dataFacility, on=['state', 'name', 'id'])
data['ind'] = np.where(data['heatInput_x'] == data['heatInput_y'], 1, 0)
data['cumSum'] = data.groupby(['state', 'name', 'id'])['ind'].cumsum()
data = data[data['cumSum'] <= 1]
data = data.drop(columns=['heatInput_y', 'ind', 'cumSum'])

data = data.rename(columns={'heatInput_x': 'heatInput'})

# now calculate heat rate and co2 rate for each plant 
data['heatRate'] = data['heatInput'] / data['load']
data['co2Rate'] = data['co2Mass'] / data['load']

# drop outliers 
data = data[data['heatRate'] < 100000]

# drop other types, these are too noisy 
data = data[data['type'] != "4 other"]
data['type'] = np.where(data['type'] == "1 coal", "Coal",
                        np.where(data['type'] == "2 oil", "Oil", "Gas"))
```

```python
data.to_csv('../output/03_context_extvalidity/plant_distribution.csv', index=False)
```

### Part C

```python
# Load PA Coal Power Plant Data  -----------------------------------------------------------
coal = pd.read_csv("../data/annual_emissions/annual-emissions-pa-coal.csv")
coal['type'] = "Coal"
coalPet = pd.read_csv("../data/annual_emissions/annual-emissions-pa-coalpet.csv")
coalPet['type'] = "Petroleum Coke"
coalRefuse = pd.read_csv("../data/annual_emissions/annual-emissions-pa-coalrefuse.csv")
coalRefuse['type'] = "Coal Refuse"

data = pd.concat([coal, coalPet, coalRefuse], ignore_index=True)
data = data.sort_values(by=['Facility ID', 'Year'])

# coal type for conversion 
coalType = pd.read_csv("../data/other/coal-type.csv")
data = pd.merge(data, coalType, how='left')

data = data.rename(columns = {'State': 'state', 'Facility Name': 'name', 'Facility ID': 'id',
 'Year': 'year', 'Heat Input (mmBtu)': 'heatInput', 'Gross Load (MWh)': 'grossLoad',
 'Steam Load (1000 lb)': 'steamLoad', 'SO2 Mass (short tons)': 'so2Mass', 'CO2 Mass (short tons)': 'co2Mass',
 'NOx Mass (short tons)': 'noxMass', 'Fuel Type': 'coalType', 'Energy': 'energy'})

# keep relevant variables and change units 
data['steamInd'] = 0
data.loc[(~data['steamLoad'].isna() & data['grossLoad'].isna()), 'steamInd'] = 1

data['energyConvert'] = np.nan
data.loc[(data['steamInd'] == 1) & (data['energy'].notnull()), 'energyConvert'] = data['energy']
data.loc[(data['steamInd'] == 1) & (data['energy'].notnull()) & (data['type'] == "1 coal"), 'energyConvert'] = 9845
data.loc[(data['steamInd'] == 1) & (data['energy'].notnull()) & (data['type'].isin(["2 oil", "3 gas"])), 'energyConvert'] = 19750

data['heatInput'] = data['heatInput'] * 10**6
data['grossLoad'] = data['grossLoad'] * 10**3
data['steamLoad'] = data['steamLoad'] * 1000 * 1/24 * data['energyConvert'] * 0.000293
data['load'] = np.where(data['grossLoad'].isna(), data['steamLoad'], data['grossLoad'])

data['co2Mass'] = data['co2Mass'] * 0.907185 * 1000
data['so2Mass'] = data['so2Mass'] * 0.907185 * 1000
data['noxMass'] = data['noxMass'] * 0.907185 * 1000

data = data[['state', 'name', 'id', 'type', 'year', 'heatInput', 'load', 'co2Mass',
             'so2Mass', 'noxMass']]
```

```python
# Filter and process data
data = data[data['year'] >= 2000]
data = data[~((data['name'] == "Gilberton Power Company") & (data['type'] == "Coal Refuse") & (data['year'] < 2003))]
data = data[~((data['name'] == "Mt. Carmel Cogeneration") & (data['type'] == "Coal Refuse") & (data['year'] == 2014))]
data = data[~((data['name'] == "Cambria Cogen") & (data['type'] == "Coal Refuse") & (data['year'] <= 2003))]
data = data[~((data['name'] == "Northampton Generating Plant") & (data['type'] == "Coal Refuse"))]
data = data[~((data['name'] == "Northampton Generating Plant") & (data['type'] == "Petroleum Coke"))]
data = data[~((data['name'] == "Seward") & (data['type'] == "Coal Refuse") & (data['year'] == 2003))]
data = data[~((data['name'] == "Gilberton Power Company") & (data['type'] == "Coal Refuse") & (data['year'] == 2003))]
data = data[~((data['name'] == "St. Nicholas Cogeneration Project") & (data['type'] == "Coal Refuse") & (data['year'] == 2009))]
```

```python
# Calculate rates
data['heatRate'] = data['heatInput'] / data['load']
data['co2Rate'] = data['co2Mass'] / data['load']
data['so2Rate'] = data['so2Mass'] / data['load']
data['noxRate'] = data['noxMass'] / data['load']

# Get minimum and maximum years
data['minyear'] = data.groupby('id')['year'].transform('min')
data['maxyear'] = data.groupby('id')['year'].transform('max')

# Retired vs. non-retired plants
data['retiredInd'] = np.where(data['maxyear'] < 2022, 1, 0)
```

```python
# Charts  ---------------------------------------------------------------------------

colors = {"Retired": "#CC6677", "Active": "#44AA99"}

# PA coal generation over the years
datayear = data[data['id'] != 50974][data['id'] != 50776].groupby('year')['load'].sum() / 10**9
dataScrubyear = data[data['id'] == 50974].groupby('year')['load'].sum() / 10**9

plt.figure(figsize=(12, 6))
plt.plot(datayear, label='Coal Generation')
plt.plot(datayear.rolling(window=3).mean(), color='darkgray', alpha=0.3, label='Rolling Mean')
plt.xlabel('year')
plt.ylabel('Coal Generation (Million MWh)')
plt.ylim(0, 150)
plt.legend(loc='upper left', fontsize=12)
plt.xticks(fontsize=14)
plt.yticks(fontsize=16)
plt.title('')
plt.savefig("../output/03_context_extvalidity/03_pa_trends.png", dpi=300)
plt.show()
```
